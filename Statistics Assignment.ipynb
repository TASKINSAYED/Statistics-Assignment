```python
#Q1) i. Write a Python function to calculate the mean of a given list of numbers. Create a function to find the median of a list of numbers.

#Code
import random

# Generate a list of 100 integers containing values between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the mean of a list of numbers
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Function to calculate the median of a list of numbers
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    midpoint = n // 2
    
    # If the list has an odd number of elements, return the middle element
    if n % 2 == 1:
        return sorted_numbers[midpoint]
    else:
        # If the list has an even number of elements, return the average of the two middle elements
        return (sorted_numbers[midpoint - 1] + sorted_numbers[midpoint]) / 2

# Test the functions with int_list
mean_value = calculate_mean(int_list)
median_value = calculate_median(int_list)

print("Mean :",mean_value, "and Median :",median_value)
```

```python
#Q1) ii. Develop a program to compute the mode of a list of integers.

#Code
import random
from collections import Counter

# Generate a list of 100 integers containing values between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]


# Function to calculate the mode of a list of numbers
def calculate_mode(numbers):
    # Count the frequency of each number in the list
    frequency = Counter(numbers)
    
    # Find the highest frequency
    max_frequency = max(frequency.values())
    
    # Find all the numbers with the highest frequency
    mode = [number for number, freq in frequency.items() if freq == max_frequency]
    
    return mode

# Test the function with int_list
mode_value = calculate_mode(int_list)
print("Mode :",mode_value)

##mode_value
```

```python
#Q1) iii. Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

#Code
import random

# Step 1: Generate a list of 100 integers between 90 and 130
values = [random.randint(90, 130) for _ in range(100)]

# Step 2: Generate a corresponding list of 100 random weights
weights = [random.randint(1, 10) for _ in range(100)]

# Step 3: Define the weighted_mean function
def weighted_mean(values, weights):
    if len(values) != len(weights):
        raise ValueError("The number of values must match the number of weights.")
    
    total_weight = sum(weights)
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    
    return weighted_sum / total_weight if total_weight != 0 else 0

# Step 4: Calculate the weighted mean of the generated lists
result = weighted_mean(values, weights)

# Output the result
print("Weighted Mean :",result)
```

```python
#Q1) iv. Write a Python function to find the geometric mean of a list of positive numbers.

#Code
import math
from functools import reduce

def geometric_mean(numbers):
    if not numbers:
        raise ValueError("The list is empty.")
    
    product = reduce(lambda x, y: x * y, numbers)
    n = len(numbers)
    return math.pow(product, 1/n)

# Example usage with the generated list
int_list = [random.randint(90, 130) for _ in range(100)]  # Using the previous list generation example
mean = geometric_mean(int_list)
print(f"The geometric mean is : {mean}")
```

```python
#Q1) vii. Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.

#Code
def trimmed_mean(numbers, trim_percentage):
    if not numbers:
        raise ValueError("The list is empty.")
    if not (0 <= trim_percentage < 50):
        raise ValueError("Trim percentage must be between 0 and 50.")

    # Sort the numbers to easily trim the smallest and largest values
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    trim_count = int(n * trim_percentage / 100)

    # Trim the smallest and largest values
    trimmed_numbers = sorted_numbers[trim_count:n - trim_count]

    if not trimmed_numbers:
        raise ValueError("Trimmed list is empty. Trim percentage too high?")
    
    # Calculate the mean of the remaining numbers
    return sum(trimmed_numbers) / len(trimmed_numbers)

# Example usage with the generated list
int_list = [random.randint(90, 130) for _ in range(100)]  # Using the previous list generation example
trim_percentage = 10  # Example: trim 10% of smallest and largest values
mean = trimmed_mean(int_list, trim_percentage)
print(f"The trimmed mean is : {mean}")
```

```python
#Q3) Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

#Code
class DiscreteRandomVariable:
    def __init__(self, probabilities, values):
        if len(probabilities) != len(values):
            raise ValueError("Probabilities and values must have the same length.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("Probabilities must sum to 1.")
        if any(p < 0 for p in probabilities):
            raise ValueError("Probabilities must be non-negative.")

        self.probabilities = np.array(probabilities)
        self.values = np.array(values)

    def expected_value(self):

        return np.sum(self.probabilities * self.values)

    def variance(self):

        mean = self.expected_value()
        return np.sum(self.probabilities * (self.values - mean) ** 2)

# Example usage
probabilities = [0.1, 0.4, 0.2, 0.3]
values = [1, 2, 3, 4]

random_variable = DiscreteRandomVariable(probabilities, values)

expected_value = random_variable.expected_value()
variance = random_variable.variance()

print(f"Expected Value: {expected_value:.2f}")
print(f"Variance: {variance:.2f}")
```

```python
#Q5) Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

#Code
import numpy as np

def generate_and_analyze_samples(distribution, **params):
    if distribution == 'binomial':
        n = params.get('n')
        p = params.get('p')
        if n is None or p is None:
            raise ValueError("For binomial distribution, parameters 'n' and 'p' are required.")
        samples = np.random.binomial(n, p, size=10000)
    
    elif distribution == 'poisson':
        lam = params.get('lam')
        if lam is None:
            raise ValueError("For Poisson distribution, parameter 'lam' is required.")
        samples = np.random.poisson(lam, size=10000)
    
    else:
        raise ValueError("Unsupported distribution. Choose 'binomial' or 'poisson'.")
    
    # Calculate mean and variance of the samples
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

binomial_mean, binomial_variance = generate_and_analyze_samples('binomial', n=10, p=0.5)
print(f"Binomial Distribution - Mean: {binomial_mean:.2f}, Variance: {binomial_variance:.2f}")

poisson_mean, poisson_variance = generate_and_analyze_samples('poisson', lam=5)
print(f"Poisson Distribution - Mean: {poisson_mean:.2f}, Variance: {poisson_variance:.2f}")
```

```python
#Q7) ii. Create a program that determines whether the columns exhibit positive skewness, negative skewness, or is approximately symmetric.

#Code
import seaborn as sns
import scipy.stats as stats

def determine_skewness_type(skewness):

    if skewness > 0:
        return "Positive Skewness (Right-Skewed)"
    elif skewness < 0:
        return "Negative Skewness (Left-Skewed)"
    else:
        return "Approximately Symmetric"

def analyze_skewness(df):

    # Calculate skewness for 'total_bill' and 'tip' columns
    skewness_total_bill = stats.skew(df['total_bill'].dropna())
    skewness_tip = stats.skew(df['tip'].dropna())
    
    # Determine skewness type
    skewness_type_total_bill = determine_skewness_type(skewness_total_bill)
    skewness_type_tip = determine_skewness_type(skewness_tip)
    
    return {
        'Total_bill': {
            'Skewness': skewness_total_bill,
            'Skewness Type': skewness_type_total_bill
        },
        'Tip': {
            'Skewness': skewness_tip,
            'Skewness Type': skewness_type_tip
        }
    }

# Load the tips dataset from Seaborn
tips = sns.load_dataset('tips')

# Analyze skewness for 'total_bill' and 'tip' columns
skewness_analysis = analyze_skewness(tips)

# Display the results
for column, results in skewness_analysis.items():
    print(f"{column} - Skewness: {results['Skewness']:.2f}, Skewness Type: {results['Skewness Type']}")
```

```python
#Q7) iii. Write a function that calculates the covariance between two columns.

#Code
import seaborn as sns
import numpy as np

def calculate_covariance(df, col1, col2):

    # Extract the two columns from the DataFrame and drop NaN values
    data = df[[col1, col2]].dropna()
    
    # Calculate the covariance matrix
    covariance_matrix = np.cov(data[col1], data[col2])
    
    # The covariance between the two columns is the (0, 1) element of the matrix
    covariance = covariance_matrix[0, 1]
    
    return covariance

# Load the tips dataset from Seaborn
tips = sns.load_dataset('tips')

# Calculate covariance between 'total_bill' and 'tip'
covariance_value = calculate_covariance(tips, 'total_bill', 'tip')

# Display the covariance
print(f"Covariance between 'Total_bill' and 'Tip': {covariance_value:.2f}")
```

```python
#Q9) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

#Code
from scipy.stats import expon

def calculate_cdf_exponential(x, rate):

    # The scale parameter is the inverse of the rate parameter ()
    scale = 1 / rate
    
    # Calculate the CDF using scipy's expon.cdf function
    cdf_values = expon.cdf(x, scale=scale)
    
    return cdf_values

rate = 0.5  # Rate parameter () of the exponential distribution
x_values = [1, 2, 3, 4, 5]  # Points at which to calculate the CDF

# Calculate the CDF for the given x values
cdf_results = calculate_cdf_exponential(x_values, rate)

# Display the results
print(f"CDF values for x = {x_values} with rate = {rate}: {cdf_results}")
```

```python
#Q10) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

#Code
from scipy.stats import poisson

def calculate_pmf_poisson(k, rate):

    # Calculate the PMF using scipy's poisson.pmf function
    pmf_values = poisson.pmf(k, mu=rate)
    
    return pmf_values

rate = 3  # Rate parameter () of the Poisson distribution
k_values = [0, 1, 2, 3, 4, 5]  # Points (number of occurrences) at which to calculate the PMF

# Calculate the PMF for the given k values
pmf_results = calculate_pmf_poisson(k_values, rate)

# Display the results
print(f"PMF values for k = {k_values} with rate = {rate}: {pmf_results}")
```

```python
#Q11) A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.
#To generate the data use the following command:
#python
#import numpy as np
# 50 purchases out of 1000 visitors
#old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors  
#new_layout = np.array([1] * 70 + [0] * 930)
#Apply z-test to find which layout is successful.

#Code
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate the number of successes and total samples
old_successes = np.sum(old_layout)
old_total = len(old_layout)

new_successes = np.sum(new_layout)
new_total = len(new_layout)

# Prepare the data for z-test
successes = np.array([old_successes, new_successes])
totals = np.array([old_total, new_total])

# Perform the z-test for proportions
z_stat, p_value = proportions_ztest(successes, totals)

# Display the results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("The difference in conversion rates is statistically significant. The new layout is more successful.")
else:
    print("The difference in conversion rates is not statistically significant. No evidence to suggest the new layout is better.")
```

```python
#Q12) A tutoring service claims that its program improves students' exam scores. A sample of students who participated in the program was taken, and their scores before and after the program were recorded.
#Use the below code to ge<erate samples of respective arrays of marks:
#python
#before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
#after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

#Code
import numpy as np
from scipy import stats

# Generate the data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the differences
differences = after_program - before_program

# Compute the mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=0)  # Population standard deviation
n = len(differences)

# Perform the z-test
# Z = (mean_diff - 0) / (std_diff / sqrt(n))
z_stat = mean_diff / (std_diff / np.sqrt(n))

# Two-tailed p-value for the z-test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Display the results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("The improvement in scores is statistically significant. The tutoring program appears to be effective.")
else:
    print("The improvement in scores is not statistically significant. No evidence to suggest the tutoring program is effective.")
```

```python
#Q15) A company is testing two different website layouts to see which one leads to higher click-through rates.
#Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.
#Use the following data:
#python
#layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
#layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

#Code
import numpy as np
from scipy import stats

def ab_test(layout_a_clicks, layout_b_clicks):

    # Convert lists to numpy arrays
    layout_a = np.array(layout_a_clicks)
    layout_b = np.array(layout_b_clicks)
    
    # Calculate sample means and standard deviations
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    
    std_a = np.std(layout_a, ddof=1)  # Sample standard deviation
    std_b = np.std(layout_b, ddof=1)  # Sample standard deviation
    
    n_a = len(layout_a)
    n_b = len(layout_b)
    
    # Calculate the t-statistic and degrees of freedom
    t_stat, p_value = stats.ttest_ind(layout_a, layout_b, equal_var=False)  # Welch's t-test
    
    # Print results
    print(f"Mean of Layout A: {mean_a:.2f}")
    print(f"Mean of Layout B: {mean_b:.2f}")
    print(f"Standard Deviation of Layout A: {std_a:.2f}")
    print(f"Standard Deviation of Layout B: {std_b:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    return t_stat, p_value

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform the A/B test
t_stat, p_value = ab_test(layout_a_clicks, layout_b_clicks)

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("There is a statistically significant difference between the two layouts.")
else:
    print("There is no statistically significant difference between the two layouts.")
```

```python
#Q16) A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the t-statistic and p-value for the treatment effect.
#Use the following data of cholestrol level:
#python
#existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
#new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

#Code
import numpy as np
from scipy import stats

def analyze_cholesterol_levels(existing_drug_levels, new_drug_levels):

    # Convert lists to numpy arrays
    existing_drug = np.array(existing_drug_levels)
    new_drug = np.array(new_drug_levels)
    
    # Calculate sample means and standard deviations
    mean_existing = np.mean(existing_drug)
    mean_new = np.mean(new_drug)
    
    std_existing = np.std(existing_drug, ddof=1)  # Sample standard deviation
    std_new = np.std(new_drug, ddof=1)  # Sample standard deviation
    
    n_existing = len(existing_drug)
    n_new = len(new_drug)
    
    # Perform the two-sample t-test
    t_stat, p_value = stats.ttest_ind(new_drug, existing_drug, equal_var=False)  # Welch's t-test
    
    # Print results
    print(f"Mean of Existing Drug: {mean_existing:.2f}")
    print(f"Mean of New Drug: {mean_new:.2f}")
    print(f"Standard Deviation of Existing Drug: {std_existing:.2f}")
    print(f"Standard Deviation of New Drug: {std_new:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    return t_stat, p_value

existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform the analysis
t_stat, p_value = analyze_cholesterol_levels(existing_drug_levels, new_drug_levels)

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("There is a statistically significant difference in cholesterol levels. The new drug is more effective.")
else:
    print("There is no statistically significant difference in cholesterol levels. The new drug is not proven to be more effective.")
```

```python
#Q17) A school district introduces an educational intervention program to improve math scores. Write a Python function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to determine if the intervention had a significant impact.
#Use the following data of test score: 
#python
#  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
#  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

#Code
import numpy as np
from scipy import stats

def analyze_intervention(pre_scores, post_scores):

    # Convert lists to numpy arrays
    pre = np.array(pre_scores)
    post = np.array(post_scores)
    
    # Perform the paired t-test
    t_stat, p_value = stats.ttest_rel(post, pre)
    
    # Calculate mean and standard deviation of the differences
    differences = post - pre
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)  # Sample standard deviation
    
    # Print results
    print(f"Mean of Differences: {mean_diff:.2f}")
    print(f"Standard Deviation of Differences: {std_diff:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    return t_stat, p_value

pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Peform the analysis
t_stat, p_value = analyze_intervention(pre_intervention_scores, post_intervention_scores)

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("The intervention had a statistically significant impact on math scores.")
else:
    print("The intervention did not have a statistically significant impact on math scores.")
```

```python
#21) A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences.
#Use the below code to generate data:
#python
#np.random.seed(0)
#age_groups = np.random.choice([ 18-30 , 31-50 , 51+', 51+'], size=30)
#voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

#Code
import numpy as np
import pandas as pd
from scipy import stats

# Generate the synthetic data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+', 'Unknown'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame from the generated data
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)
print("Expected Frequencies Table:")
print(expected)

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nThere is a statistically significant association between age groups and voter preferences.")
else:
    print("\nThere is no statistically significant association between age groups and voter preferences.")
```
